## Elevation by Zone

```python
df_sorted = df.sort_values('elevation_m')

fig, ax = plt.subplots(figsize=(10, 5))
colors = ['#c62828' if e < 6 else '#ef6c00' if e < 10 else '#f9a825' if e < 13 else '#388e3c'
          for e in df_sorted.elevation_m]
ax.barh(df_sorted.zone_name, df_sorted.elevation_m, color=colors)
ax.set_xlabel('Mean Elevation (m)')
ax.set_title('Chennai GCC Zones — Real SRTM Elevation (lowest = most flood-prone)')
ax.axvline(6, color='red', ls='--', lw=0.8, alpha=0.5, label='<6m threshold')
ax.legend(fontsize=8)
plt.tight_layout()
plt.show()

print('\nSource: USGS SRTM 30m DEM, zonal statistics extracted in QGIS')
print('Boundaries: TNGIS official GCC 15-zone layer')
```
